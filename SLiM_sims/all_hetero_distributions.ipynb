```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Neutality

```python
neutrality = pd.read_csv('/home/miguel/Desktop/HonoursProject/SLiMsimulations/Selection_sims/Plots/neutrality.txt')
neutrality = neutrality.values.tolist()
neutrality = np.array(neutrality)
```

```python
plt.hist(neutrality, bins = 50, color = 'forestgreen', edgecolor = 'darkgreen', alpha = 0.8)
plt.xlabel("Proportion of heterozygosity remaining")
plt.ylabel("Number of replicates")
plt.title('Distribution of Heterozygosity under neutrality')
plt.savefig(r'/home/miguel/Desktop/HonoursProject/SLiMsimulations/Selection_sims/Plots/neutrality.png')
```

```python
print(sum(neutrality)/len(neutrality))
print(np.percentile(neutrality, 95))
```
