```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Neutrality

```python
neutral_overlap = pd.read_csv('/home/miguel/Desktop/HonoursProject/SLiMsimulations/tests2/tract_overlap.txt')
neutral_overlap = neutral_overlap.values.tolist()
neutral_overlap = np.array(neutral_overlap)
```

```python
plt.hist(neutral_overlap, bins = 70, color = 'forestgreen', alpha = 0.8, edgecolor = 'darkgreen')
plt.xlabel("Tract overlap under neutrality")
plt.ylabel("Number of replicates")
plt.title('Distribution of Heterozygosity tract overlap under neutrality')
plt.savefig(r'/home/miguel/Desktop/HonoursProject/SLiMsimulations/overlaps/Tract_overlap_neutrality.png')
```

```python
print(np.percentile(neutral_overlap, 95))
print(sum(neutral_overlap)/len(neutral_overlap))
print(max(neutral_overlap))
```
